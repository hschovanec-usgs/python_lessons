## Line Plot

```python
## Example 1
x = np.arange(0, 2*np.pi, np.pi/100)
y = np.sin(x)

plt.figure(1)
plt.plot(x, y)
plt.title('Example 1')
plt.show()
```

## Multiple Line Plot

```python
## Example 2
x1 = np.linspace(-2*np.pi, 2*np.pi)
y1 = np.sin(x1)
y2 = np.cos(x1)

plt.figure(2)
plt.plot(x1, y1, x1, y2)
plt.title('Example 2')
plt.show()
```

## Line Plot from Matrix

```python
## Example 3
Y = np.random.rand(4, 4)

plt.figure(3)
plt.plot(Y)
plt.title('Example 3')
plt.show()
```

## Line Styling

```python
## Example 4
x = np.arange(0, 2*np.pi, np.pi/100)
y1 = np.sin(x)
y2 = np.sin(x-0.25)
y3 = np.sin(x-0.5)

plt.figure(4)
plt.plot(x, y1, x, y2, '--', x, y3, ':')
plt.title('Example 4')
plt.show()
```

```python
## Example 5
x = np.arange(0, 2*np.pi, np.pi/10)
y1 = np.sin(x)
y2 = np.sin(x-0.25)
y3 = np.sin(x-0.5)

plt.figure(5)
plt.plot(x, y1, 'g', x, y2, 'b--o', x, y3, 'c*')
plt.title('Example 5')
plt.show()
```

## Markers at Specific Data Points

```python
## Example 6
x = np.linspace(0, 10)
y = np.sin(x)

plt.figure(6)
plt.plot(x,y,'-o', markevery=list(np.arange(0, len(y), 4)))
plt.title('Example 6')
plt.show()
```

```python
## Example 7
x = np.arange(-np.pi, np.pi, np.pi/10)
y = np.tan(np.sin(x)) - np.sin(np.tan(x));

plt.figure(7)
plt.plot(x,y,'--gs', linewidth=2, markersize=10,
         markeredgecolor='b', markerfacecolor=[0.5,0.5,0.5])
plt.title('Example 7')
plt.show()
```

## Labels

```python
## Example 8
x = np.linspace(0, 10, 150)
y = np.cos(5*x)

plt.figure(8)
plt.plot(x, y, color=[0,0.7,0.9])
plt.title('Example 8: 2-D Line Plot')
plt.xlabel('x')
plt.ylabel('cos(5x)')
plt.show()
```

## Plotting time

```python
## Example 9
x = [datetime.datetime.now() + datetime.timedelta(seconds=int(i)) for i in np.arange(0, 3*60, 30)]
y = np.random.rand(6)

fig, ax = plt.subplots()
ax.plot(x,y)
myFmt = mdates.DateFormatter('%M:%S')
ax.xaxis.set_major_formatter(myFmt)
ax.set_title('Example 9')
plt.show()
```

## Multiple Axes

```python
## Example 10
x = np.linspace(0,3)
y1 = np.sin(5*x)
y2 = np.sin(15*x)

f, (ax1, ax2) = plt.subplots(2, 1, sharex=True)
ax1.plot(x1, y1)
ax1.set_title('Top Plot')
ax1.set_ylabel('sin(5x)')

ax2.scatter(x1, y2)
ax2.set_title('Bottom Plot')
ax2.set_ylabel('sin(15x)')
plt.show()
```

## Plotting Circles

```python
## Example 10
radius = 2
center_x = 4
center_y = 3
theta = np.linspace(0, 2*np.pi)
x = radius*np.cos(theta) + center_x
y = radius*np.sin(theta) + center_y

plt.figure(10, figsize=(5, 5))
plt.plot(x, y)
plt.title('Example 10')
plt.show()
```
